# Scatter plot

```python
subset_df = df.head(200)

plt.figure(figsize=(10, 5))  
sns.set(style='dark')  

sns.scatterplot(x=subset_df['TSH_Level'], y=subset_df['Nodule_Size'], hue=subset_df['Thyroid_Cancer_Risk'], palette="coolwarm")

plt.title("Scatter Plot of TSH Level vs Nodule Size (Subset of 200 Records)")  
plt.xlabel("TSH Level")  
plt.ylabel("Nodule Size")  

plt.show() 
```

# Heatmap

```python
subset_df = df.head(200)

correlation_matrix = subset_df[['Age', 'TSH_Level', 'T3_Level', 'T4_Level', 'Nodule_Size']].corr()

plt.figure(figsize=(6, 4))  
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1)

plt.title("Correlation Heatmap (Subset of 200 Records)")  
plt.show()  
```
